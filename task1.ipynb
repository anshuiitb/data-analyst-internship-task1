```python
# Importing Libraries
import pandas as pd
import numpy as np
```

```python
import pandas as pd

df = pd.read_csv("/content/marketing_campaign.csv", sep='\t')# storing csv in pandas dataframe
```

```python
df.head() # top 5 rows
```

```python
df.isnull().sum() #missing values count of each column
```

```python
df['Income'].fillna(df['Income'].mean(), inplace=True) #replacing missing values in column named income by average income
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')  # Convert all column names to lowercase and replace spaces with underscores for consistency
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop_duplicates(inplace=True) #removing duplicate rows
```

```python
df['education'].unique() # all unique values of column 'education'
```

```python
df['education'] = df['education'].str.strip().str.title()  # Trim whitespace and capitalize
df['education'] = df['education'].replace({'2n Cycle': 'Master'})  # replacing '2n cycle' in education's column to 'Master'
```

```python
df['marital_status'].unique()# all unique values of column 'martial_status'
```

```python
df['marital_status']= df['marital_status'].str.strip().str.title()  # Trim whitespace and capitalize
df['marital_status'] = df['marital_status'].replace({'Alone': 'Single'})
```

```python
df['dt_customer'] = pd.to_datetime(df['dt_customer'], format='%d-%m-%Y')  # Standardize to datetime
```

```python
#making sure that year_birth is if int datatype and income is of datatype float
df['year_birth'] = df['year_birth'].astype(int)
df['income'] = df['income'].astype(float)
```

**Outlier and Validation**

```python
from scipy import stats  # For z-scores
numerical_cols = ['income', 'mntwines', 'mntfruits', 'mntmeatproducts', 'mntfishproducts', 'mntsweetproducts', 'mntgoldprods']
z_scores = np.abs(stats.zscore(df[numerical_cols].select_dtypes(include=[np.number])))
z_scores
```

```python
outliers = (z_scores > 3).any(axis=1)
print(f"Number of rows with potential outliers: {outliers.sum()}")
```

```python
df['income'] = np.where(df['income'] > 200000, 200000, df['income'])
```

```python
# purchase_cols = ['mntwines', 'mntfruits', 'mntmeatproducts', 'mntfishproducts', 'mntsweetproducts', 'mntgoldprods']
# for col in purchase_cols:
#     df[col] = df[col].apply(lambda x: 0 if x < 0 else x)

purchase_cols = ['mntwines', 'mntfruits', 'mntmeatproducts', 'mntfishproducts', 'mntsweetproducts', 'mntgoldprods']
for col in purchase_cols:
    df[col] = df[col].clip(lower=0)  # More efficient than apply
```

```python
# Validate year_birth (set future years to NaN; current year is 2025)
df['year_birth'] = df['year_birth'].apply(lambda x: x if x <= 2025 else np.nan)
```

```python
# Optional: Calculate age and flag extreme ages (e.g., >120 as invalid)
df['age'] = 2025 - df['year_birth']
df['age'] = df['age'].apply(lambda x: np.nan if x > 120 else x)  # Justify: Unrealistic ages likely errors
```

```python
# for new missing values introduced
print(df.isnull().sum())
```

```python
df = df.dropna(subset=['age'])
print("Rows dropped; new shape:", df.shape)
```

```python
# box-plot for income and age visualization
import matplotlib.pyplot as plt
df.boxplot(column=['income', 'age'])
plt.title('Boxplot for Outliers')
plt.show()
```

```python
# Re-check for new missing values introduced
print(df.isnull().sum())
```
